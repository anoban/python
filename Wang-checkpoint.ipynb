```python
def complement(sequence: str) -> str:
    """
    Gives the complementary sequence of a DNA!
    Can only handle A,T,C,G bases as of now!
    """
    LOOKUP = {'A':'T', 'T':'A', 'C':'G', 'G':'C'}
    return "".join((LOOKUP.get(s, 'N') for s in sequence))
```

```python
complement(kword) in salim
```

```python
complement(kword)[::-1] in salim
```

```python
complement(kword[::-1]) in salim
```

```python
reversed_complementary: str = "".join([f">{t}\n{complement(s)[::-1]}\n\n" for (t, s) in zip(titles, sequences)])
```

```python
print(reversed_complementary)
```

```python
with open(r"./Wang-Rev-Compl.fasta", mode = 'w') as fp:
    fp.write(reversed_complementary)
```
