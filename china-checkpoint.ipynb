```python
os.listdir("./data/seqs/edited/")
```

```python
identifiers = []
fastas = []

for fa in os.listdir("./data/seqs/edited/"):
    if fa.endswith(".fa"):
        with open("./data/seqs/edited/" + fa, 'r') as fp:
            text = fp.read()
            identifiers.append(text.splitlines()[0])
            fastas.append("".join(text.splitlines()[1:]))
        
```

```python
identifiers
```

```python
antis = ["".join([lookup.get(n) for n in seq]) for seq in fastas]
```

```python
antis
```

```python
with open("./data/seqs/anti.fasta", 'w') as fp:
    for ident, anti in zip(identifiers, antis):
        fp.write(f"{ident}\n{anti}\n")
```
