```python
import numpy as np
np.random.seed(2024 - 7 - 12)
import matplotlib.pyplot as plt
```

```python
x = np.arange(-5, 5.01, step = 0.02)
y = x
```

```python
x.size, y.size
```

### ___$f(x,~y) = x^2 + y^2$___

```python
z = np.square(x) + np.square(y)
```

```python
fig = plt.figure(figsize = (10, 10))
axes = fig.add_subplot(projection = "3d")
axes.plot(x, y, z)
plt.show()
```
