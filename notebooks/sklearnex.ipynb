```python
from sklearnex.neighbors import KNeighborsClassifier
from sklearnex.linear_model import LinearRegression, LogisticRegression
from sklearn.preprocessing import StandardScaler, LabelEncoder, LabelBinarizer
from sklearnex.model_selection import train_test_split
from sklearn.metrics import accuracy_score   # vanila sklearn, sklearnex doesn't have accuracy_score
import numpy as np
import pandas as pd
from scipy.io import arff
```

```python
df = pd.DataFrame(arff.loadarff("./phpCsX3fx.arff")[0])
```

```python
df.columns
```

```python
df.Class = df.Class.apply(lambda x: np.int64(x))
```

```python
df
```

```python
df.shape
```

```python
x = df.iloc[:, : df.shape[1] - 1]
y = df.iloc[:, df.shape[1] - 1]
```

```python
sscaler = StandardScaler()
x = sscaler.fit_transform(x)
```

```python
train_x, test_x, train_y, test_y = train_test_split(x, y, test_size = 0.3)
```

```python
knn = KNeighborsClassifier(n_neighbors = 10, n_jobs = 8, leaf_size = 10, p = 10)
```

```python
knn.fit(train_x, train_y)
```

```python
knn.predict(test_x)
```

```python
lreg = LinearRegression(n_jobs = 8)
```

```python
lreg.fit(train_x, train_y)
```

```python
lreg.coef_
```

```python
lreg.predict(test_x)
```

```python
rice = pd.DataFrame(arff.loadarff("./Rice_Cammeo_Osmancik.arff")[0])
```

```python
rice.columns
```

```python
rice_x = sscaler.fit_transform(rice.iloc[:, : rice.shape[1] - 1])
```

```python
rice.Class.unique()
```

```python
rice_y = LabelEncoder().fit_transform(rice.Class)
```

```python
rice_train_x, rice_test_x, rice_train_y, rice_test_y = train_test_split(rice_x, rice_y, test_size = 0.3)
```

```python
preds = KNeighborsClassifier(n_jobs = 16).fit(rice_train_x, rice_train_y).predict(rice_test_x)
```

```python
# wow!

accuracy_score(preds, rice_test_y)
```

```python
preds = LinearRegression(n_jobs = 16).fit(rice_train_x, rice_train_y).predict(rice_test_x)
```

```python
accuracy_score(preds, rice_test_y)
```

```python
preds = LogisticRegression(max_iter = 1000, n_jobs = 16).fit(rice_train_x, rice_train_y).predict(rice_test_x)
```

```python
# that's a little better

accuracy_score(preds, rice_test_y)
```

```python
bean = pd.read_csv("./Dry_Bean_Dataset.csv")
```

```python
bean.shape
```

```python
bean_x = StandardScaler().fit_transform(bean.iloc[:, : bean.shape[1] - 1])
```

```python
bean_y = LabelEncoder().fit_transform(bean.iloc[:, 16])
```

```python
np.unique(bean_y)
```

```python
train_x, test_x, train_y, test_y = train_test_split(bean_x, bean_y)
```

```python
preds = LogisticRegression(n_jobs = 16, max_iter = 1000).fit(train_x, train_y).predict(test_x)
```

```python
# that truly is impressive.

accuracy_score(preds, test_y)
```
