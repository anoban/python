## ___%timeit___

```python
# the statement that needs to be timed must be in the same line as the %timeit command!

%timeit x: list[int] = [i * i for i in range(10_000)]
```

```python
x: int = 0;
y: list[float] = [i / 10 + sum(y) for i in range(1, 10_000)]

%timeit z: list[float] = [(i / 10 + sum(y)) * i for i in range(1, 10_000)]
```

## ___%%timeit___

```python
%%timeit

# %timeit for cells, since it is a cell magic command, it needs to be the first thing in a cell
# won't work otherwise

def square(number: int | float) -> float:
    """
    Returns the square of a number.
    number: int | float
    returns a float
    """
    if (not isinstance(number, int)) and (not isinstance(number, float)):
        raise TypeError("Only int or float are accepted as inputs!")
    return number * number


[square(b) for b in range(1, 20_000, 2)]
```

```python
# this won't work! See

%%timeit

[square(b) for b in range(1, 20_000, 2)]
```

```python
square("a")
```

```python
square([])
```
