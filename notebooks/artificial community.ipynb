```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(2024 - 1 - 17)
```

# ___Generating Pictures For Basic Statistical Analyses On Artificial Communities___
----------------------

```python
# our community needs to have 100 quardrat areas (10 x 10)
# each quadrat is of 1.5 m x 1.5 m in dimension
# the community comprises of 4 different species, pictorially represented in the picture using 4 different shapes (markers)
# species A, B, C and D

# let's make species A and B clustered
# C uniform 
# D completely random

Dx = np.abs(np.random.standard_normal(size = 100)) * 10
Dy = np.abs(np.random.standard_normal(size = 100)) * 10

Cx = np.arange(1, 11, step = 1) - 0.5
```

```python
Ax = np.append(np.abs(np.random.normal(loc = .10, scale = 0.25, size = 50) * 10),
               np.abs(np.random.normal(loc = .80, scale = 0.05, size = 50) * 10))
Ay = Ax + np.random.normal(loc = 1, scale = 1, size = 100)  # introduce some noise
```

```python
Bx = np.abs(np.random.normal(loc = 6, scale = 2, size = 100))
By = np.random.normal(loc = 5, scale = 1.25, size = 100)
```

```python
fig, axes = plt.subplots()
fig.set_size_inches(20, 20)
axes.set_xlim(0, 10)
axes.set_xticks(np.arange(0, 10, 1))
axes.set_ylim(0, 10)
axes.set_yticks(np.arange(0, 10, 1))
axes.set_axisbelow(True)
axes.grid(which = "both")

axes.scatter(Dx, Dy, color = "red", marker = "^", s = 100, alpha = 0.75, label = "D")
for row in range(1, 10):
    axes.scatter(Cx, np.repeat(row, 10) - 0.5, color = "green", marker = "s", s = 100, alpha = 0.75)
axes.scatter(Cx, np.repeat(10, 10) - 0.5, color = "green", marker = "s", s = 100, alpha = 0.75, label = "C")
axes.scatter(Ax, Ay, color = "blue", marker = "o", s = 100, alpha = 0.75, label = "A")
axes.scatter(Bx, By, color = "white", marker = "o", edgecolor = "black", s = 100, label = "B", alpha = 0.5)
axes.legend(loc = "upper left")
plt.tight_layout()
plt.savefig("community.png", format = "png", dpi = 500)
```
